##Regressão Linear e Logistica

###Bibliotecas

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression, LogisticRegression
from yellowbrick.regressor import ResidualsPlot
import statsmodels.formula.api as sm
```

###Regressão Linear Simples

```python
base = pd.read_csv('cars.csv')
base.shape
```

```python
base.head()
```

```python
base = base.drop(['Unnamed: 0'], axis= 1)
base.head()
```

```python
# DEFINIÇÃO DAS VARIÁVEIS X E Y, X DISTANCIA É A VARIAVEL INDEPENDENTE E Y VELOCIDADE É A VARIAVEL DEPENDENTE

X = base.iloc[:, 1].values
y = base.iloc[:, 0].values
X
```

```python
#CALCULO DA CORRELAÇÃO ENTRE X E Y
correlacao = np.corrcoef(X, y)
correlacao
```

```python
#FORMATO DE MATRIZ COM UMA COLUNA A MAIS
X = X.reshape(-1, 1)

#CRIAÇÃO DO MODELO
modelo = LinearRegression()
modelo.fit(X, y)
```

```python
modelo.intercept_
```

```python
modelo.coef_
```

```python
plt.scatter(X, y)
plt.plot(X, modelo.predict(X), color='red')
```

```python
#Qual velocidade se leveu 22 pés para parar

modelo.intercept_ + modelo.coef_ * 22
```

```python
#Previsão utilizando função da sklearn
modelo.predict([[22]])
```

```python
#Gráfico para visualizar os residuais
visualizador = ResidualsPlot(modelo)
visualizador.fit(X, y)
visualizador.poof()
```
